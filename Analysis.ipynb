New approach to truly normalise difference. (sim - target)/max(sim - target)

```python
# plot each set of obs with a target
targetObs = b.targets()
# find the normalised difference between each value and the target
for obs in obsSets:
    # find normalised differences
    diffNHex10 = (obs10[obs[0]] - targetObs[obs[0]])/np.max(np.abs(obs10[obs[0]] - targetObs[obs[0]]))
    diffNHex14 = (obs14[obs[0]] - targetObs[obs[0]])/np.max(np.abs(obs14[obs[0]] - targetObs[obs[0]]))
    difftrop10 = (obs10[obs[1]] - targetObs[obs[1]])/np.max(np.abs(obs10[obs[1]] - targetObs[obs[1]]))
    difftrop14 = (obs14[obs[1]] - targetObs[obs[1]])/np.max(np.abs(obs14[obs[1]] - targetObs[obs[1]]))
    diffSHex10 = (obs10[obs[2]] - targetObs[obs[2]])/np.max(np.abs(obs10[obs[2]] - targetObs[obs[2]]))
    diffSHex14 = (obs14[obs[2]] - targetObs[obs[2]])/np.max(np.abs(obs14[obs[2]] - targetObs[obs[2]]))
    
    # plot bar chart for NH region
    fig, ax = plt.subplots(figsize=(15,7))
    ax.bar(iterations-0.2, diffNHex10, width=0.4, label='10 params')
    ax.bar(iterations+0.2, diffNHex14, width=0.4, label='14 params')
    ax.axvline(x=25-0.2, color='xkcd:grey', label='10 params')
    ax.axvline(x=18+0.2, linestyle='--', color='xkcd:grey', label='14 params')
    ax.set_ylabel('Normalised difference (simObs - targetObs)')
    ax.set_ylim(-1,1)
    ax.set_title(f'{obs[0]}')
    ax.legend()
    plt.savefig(f'{obs[0]}Difference')
    plt.show()
    
    # plot bar chart for trop region
    fig, ax = plt.subplots(figsize=(15,7))
    ax.bar(iterations-0.2, difftrop10, width=0.4, label='10 params')
    ax.bar(iterations+0.2, difftrop14, width=0.4, label='14 params')
    ax.axvline(x=25-0.2, color='xkcd:grey', label='10 params')
    ax.axvline(x=18+0.2, linestyle='--', color='xkcd:grey', label='14 params')
    ax.set_ylabel('Normalised difference (simObs - targetObs)')
    ax.set_ylim(-1,1)
    ax.set_title(f'{obs[1]}')
    ax.legend()
    plt.savefig(f'{obs[1]}Difference')
    plt.show()
    
    # plot bar chart for SH region
    fig, ax = plt.subplots(figsize=(15,7))
    ax.bar(iterations-0.2, diffSHex10, width=0.4, label='10 params')
    ax.bar(iterations+0.2, diffSHex14, width=0.4, label='14 params')
    ax.axvline(x=25-0.2, color='xkcd:grey', label='10 params')
    ax.axvline(x=18+0.2, linestyle='--', color='xkcd:grey', label='14 params')
    ax.set_xlabel('Iterations')
    ax.set_ylabel('Normalised difference (simObs - targetObs)')
    ax.set_ylim(-1,1)
    ax.set_title(f'{obs[2]}')
    ax.legend()
    plt.savefig(f'{obs[2]}Difference')
    plt.show()
    
diffnet10 = (obs10[net] - targetObs[net])/np.max(np.abs(obs10[net] - targetObs[net]))
diffnet14 = (obs14[net] - targetObs[net])/np.max(np.abs(obs14[net] - targetObs[net]))  
    
# plot bar chart for SH region
fig, ax = plt.subplots(figsize=(15,7))
ax.bar(iterations-0.2, diffnet10, width=0.4, label='10 params')
ax.bar(iterations+0.2, diffnet14, width=0.4, label='14 params')
ax.axvline(x=25-0.2, color='xkcd:grey', label='10 params')
ax.axvline(x=18+0.2, linestyle='--', color='xkcd:grey', label='14 params')
ax.set_xlabel('Iterations')
ax.set_ylabel('Normalised difference (simObs - targetObs)')
ax.set_ylim(-1,1)
ax.set_title('net')
ax.legend()
plt.savefig('netDifference')
plt.show()
```
